```python
import zipfile
import os

# Path to your downloaded ZIP file
zip_path = '/content/archive (2).zip'
extract_path = 'FER_dataset/'

# Unzipping
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Dataset unzipped successfully!")
```

```python
from tensorflow.keras.applications import MobileNet # Import MobileNet
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Flatten, Dense

base_model = MobileNet(input_shape=(224, 224, 3), include_top=False)

for layer in base_model.layers:
  layer.trainable = False

x = Flatten()(base_model.output)
x = Dense(units=7, activation='softmax')(x)

# creating our model.
model = Model(base_model.input, x)
```

```python
from tensorflow.keras.losses import categorical_crossentropy  # Import categorical_crossentropy

model.compile(optimizer='adam', loss=categorical_crossentropy, metrics=['accuracy'])
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator # Import ImageDataGenerator

train_datagen = ImageDataGenerator(
     zoom_range = 0.2,
     shear_range = 0.2,
     horizontal_flip=True,
     rescale = 1./255
)

train_data = train_datagen.flow_from_directory(directory= "/content/FER_dataset/train",
                                               target_size=(224,224),
                                               batch_size=32,
                                  )


train_data.class_indices
```

```python
val_datagen = ImageDataGenerator(rescale = 1./255 )

val_data = val_datagen.flow_from_directory(directory= "/content/FER_dataset/test",
                                           target_size=(224,224),
                                           batch_size=32,
                                  )
```

```python
t_img , label = train_data.__next__() # Use __next__() instead of next()
#-----------------------------------------------------------------------------
# function when called will prot the images
def plotImages(img_arr, label):
  """
  input  :- images array
  output :- plots the images
  """
  count = 0
  for im, l in zip(img_arr,label) :
    plt.imshow(im)
    plt.title(im.shape)
    plt.axis = False # This line was also incorrect, it should be plt.axis('off')
    plt.show()

    count += 1
    if count == 10:
      break

#-----------------------------------------------------------------------------
# function call to plot the images
plotImages(t_img, label)
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping

# early stopping
es = EarlyStopping(monitor='val_accuracy', min_delta= 0.01 , patience= 5, verbose= 1, mode='auto')

# model check point
# Change the filepath to end in .keras
mc = ModelCheckpoint(filepath="best_model.keras", monitor= 'val_accuracy', verbose= 1, save_best_only= True, mode = 'auto')

# puting call back in a list
call_back = [es, mc]
```

```python
hist = model.fit(train_data,
                   steps_per_epoch=10,
                   epochs=30,
                   validation_data=val_data,
                   validation_steps=8,
                   callbacks=[es, mc])
```

```python
import cv2
from tensorflow.keras.preprocessing.image import img_to_array
import numpy as np
import matplotlib.pyplot as plt

# Load Haar Cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load the image
path = "/content/angry.jpg"
img = cv2.imread(path)

# Detect faces in the color image
faces = face_cascade.detectMultiScale(img, scaleFactor=1.1, minNeighbors=5)

# Ensure at least one face is detected
if len(faces) > 0:
    x, y, w, h = faces[0]  # Take the first detected face
    face = img[y:y+h, x:x+w]  # Extract face from the color image
    face_resized = cv2.resize(face, (224, 224))  # Match your target size

    # Normalize and reshape
    face_array = img_to_array(face_resized) / 255.0
    input_arr = np.expand_dims(face_array, axis=0)

    # Predict
    pred = np.argmax(model.predict(input_arr))
    print(f"The image is of {op[pred]}")

    # Display the cropped face
    plt.imshow(face_resized[:, :, ::-1])  # Display with correct color order (RGB)
    plt.title("Cropped Face")
    plt.show()
else:
    print("No face detected in the image.")
```
